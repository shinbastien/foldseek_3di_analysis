### Analyze what SCOPe folds/superfamilies/families CPs belong to

```python
import pickle
```

### Load unique CP structure names

```python

fp = '/home/ubuntu/CIRPIN/scope40/unique_CPs/unique_CPs_scope40.txt'

with open(fp, 'r') as f:
    lines = [line.strip() for line in f]
```

### Load dictionary to get sccs (i.e. a.1.1.10) for a given domain

```python
dict_fp = '/home/ubuntu/CIRPIN/scope40/unique_CPs/scope40_sid_sccs_dictionary/dir.des.scope.2.08-2023-01-06_sid_sccs_dictionary.pkl'
with open(dict_fp, 'rb') as f:
    scope_sccs_dict = pickle.load(f)
```

```python
sccs_counts = {}

for domain in lines:
    sccs = scope_sccs_dict[domain]
    if '.'.join(sccs.split('.')[:2]) == 'd.58':
        print(domain)
    sccs_counts[sccs] = sccs_counts.get(sccs, 0) + 1

    
```

```python
super_fam_counts = {}

for key, value in sccs_counts.items():
    # Split on '.' and take first 3 parts
    truncated_key = '.'.join(key.split('.')[:3])
    #print(truncated_key)
    super_fam_counts[truncated_key] = super_fam_counts.get(truncated_key,0)+value
```

```python
fold_counts = {}

for key, value in super_fam_counts.items():
    # Split on '.' and take first 3 parts
    truncated_key = '.'.join(key.split('.')[:2])
    #print(truncated_key)
    
    fold_counts[truncated_key] = fold_counts.get(truncated_key,0)+value
```

```python
fold_counts
```

```python
beta_fold_proteins_cps = {}
for key, value in fold_counts.items():
    if key.startswith('b'):
        beta_fold_proteins_cps[key] = value

all_beta_folds = []
for key, value in scope_sccs_dict.items():
    if value.startswith('b'):
        b_fold = '.'.join(value.split('.')[:2])
        if b_fold not in all_beta_folds:
            all_beta_folds.append(b_fold)
```

```python
alpha_fold_proteins_cps = {}
for key, value in fold_counts.items():
    if key.startswith('a'):
        alpha_fold_proteins_cps[key] = value

all_alpha_folds = []
for key, value in scope_sccs_dict.items():
    if value.startswith('a'):
        a_fold = '.'.join(value.split('.')[:2])
        if a_fold not in all_alpha_folds:
            all_alpha_folds.append(a_fold)
```

```python
alpha_beta_fold_proteins_cps = {}
for key, value in fold_counts.items():
    if key.startswith('c'):
        alpha_beta_fold_proteins_cps[key] = value

all_alpha_beta_folds = []
for key, value in scope_sccs_dict.items():
    if value.startswith('c'):
        ab_fold = '.'.join(value.split('.')[:2])
        if ab_fold not in all_alpha_beta_folds:
            all_alpha_beta_folds.append(ab_fold)
```

```python
d_alpha_beta_fold_proteins_cps = {}
for key, value in fold_counts.items():
    if key.startswith('d'):
        d_alpha_beta_fold_proteins_cps[key] = value

all_d_alpha_beta_folds = []
for key, value in scope_sccs_dict.items():
    if value.startswith('d'):
        ab_fold = '.'.join(value.split('.')[:2])
        if ab_fold not in all_d_alpha_beta_folds:
            all_d_alpha_beta_folds.append(ab_fold)
```

```python
#beta_fold_proteins_cps
sorted_beta_fold_cps = dict(sorted(beta_fold_proteins_cps.items(), key=lambda x: int(x[0].split('.')[1])))
```

```python
#alpha_beta_fold_proteins_cps
sorted_alpha_beta_fold_cps = dict(sorted(alpha_beta_fold_proteins_cps.items(), key=lambda x: int(x[0].split('.')[1])))
```

```python
#alpha_beta_fold_proteins_cps
sorted_d_alpha_beta_fold_cps = dict(sorted(d_alpha_beta_fold_proteins_cps.items(), key=lambda x: int(x[0].split('.')[1])))
```

```python
#alpha_beta_fold_proteins_cps
sorted_alpha_fold_cps = dict(sorted(alpha_fold_proteins_cps.items(), key=lambda x: int(x[0].split('.')[1])))
```

```python
# Get the max number
max_num = len(all_beta_folds)

# Fill in missing b.X values
filled_beta_fold_cps = {}
for i in range(1, max_num + 1):
    key = f'b.{i}'
    filled_beta_fold_cps[key] = sorted_beta_fold_cps.get(key, 0)
```

```python
# Extract keys and values
beta_folds = list(sorted_beta_fold_cps.keys())
frequencies = list(sorted_beta_fold_cps.values())

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 6))

# Create bar chart
bars = ax.bar(beta_folds, frequencies, color='#D2F8D1', edgecolor='black', linewidth=0.5)

# Set labels and title
ax.set_xlabel('SCOPe Class, Fold', fontsize=12, fontweight='bold')
ax.set_ylabel('Frequency', fontsize=12, fontweight='bold')
ax.set_title('SCOPe Class Beta, Fold Frequency Distribution', fontsize=14, fontweight='bold')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=16)
plt.yticks(fontsize=16)

# Set y-axis limits
ax.set_ylim(0, max(frequencies) * 1.1)

# Add grid
ax.grid(axis='y', alpha=0.3, linestyle='--')

# Adjust layout to prevent label cutoff
plt.tight_layout()
plt.savefig('/home/ubuntu/scope40/scope40_sid_sccs_dictionary/scope_beta_fold_dist.png')
# Display the plot
plt.show()
```
