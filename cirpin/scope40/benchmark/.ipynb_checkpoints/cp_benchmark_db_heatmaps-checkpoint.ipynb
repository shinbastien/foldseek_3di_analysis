```python
import os
import numpy as np
import matplotlib.pyplot as plt
import re

# PDB IDs
y_axis_ids = ['d1qasa2', 'd3cnaa_', 'd1puja_', 'd1qdma1', 'd1cida1', 
              'd1p8ja1', 'd1booa_', 'd2dk8a1', 'd1vhra_', 'd1mkza1', 'd1iu9a_']

x_axis_ids = ['d1rsya1', 'd2pela_', 'd1zbda_', 'd2gtga_', 'd1dcea2', 
              'd1ca1a2', 'd1g38a1', 'd1qzya1', 'd1phra_', 'd1jbwa1', 'd1h0ra_']

# Initialize score matrix
scores = np.zeros((len(y_axis_ids), len(x_axis_ids)))

# Run TM-align for all pairs
for i, pdb_q in enumerate(y_axis_ids):
    for j, pdb_t in enumerate(x_axis_ids):
        print(f"Aligning {pdb_q} vs {pdb_t}...")
        
        # Construct full paths to PDB files
        pdb_q_path = f"/home/ubuntu/CIRPIN/scope40/benchmark/all_pdbs/{pdb_q}.pdb"
        pdb_t_path = f"/home/ubuntu/CIRPIN/scope40/benchmark/all_pdbs/{pdb_t}.pdb"
        
        # Run TM-align
        output = os.popen(f'/home/ubuntu/TM_align/TMalign {pdb_q_path} {pdb_t_path}').read()
        
        # Parse TM-scores from output
        # TM-align outputs two TM-scores normalized by different lengths
        matches = re.findall(r'TM-score=\s*(\d+\.\d+)', output)
        if len(matches) >= 2:
            tm_score1 = float(matches[0])
            tm_score2 = float(matches[1])
            tm_score = min(tm_score1, tm_score2)
            scores[i, j] = tm_score
            print(f"  Scores: {tm_score1:.4f}, {tm_score2:.4f} -> Min: {tm_score:.4f}")
        elif len(matches) == 1:
            tm_score = float(matches[0])
            scores[i, j] = tm_score
            print(f"  Score: {tm_score:.4f}")
        else:
            print(f"  Warning: Could not parse TM-score for {pdb_q} vs {pdb_t}")
            scores[i, j] = np.nan
```

```python
tm_align_scores = scores
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Create a custom colormap with white at 0.8
# Define the colors: dark (below 0.8) -> white (at 0.8) -> hot (above 0.8)
colors_below = plt.cm.Blues_r(np.linspace(0, 1, 128))  # Blues reversed for below
colors_above = plt.cm.Reds(np.linspace(0, 1, 128))      # Reds for above

# Combine the colormaps
all_colors = np.vstack((colors_below, colors_above))
custom_cmap = mcolors.LinearSegmentedColormap.from_list('custom_diverging', all_colors)

# Create a normalization that centers white at 0.5
divnorm = mcolors.TwoSlopeNorm(vmin=0.17, vcenter=0.5, vmax=1.0)

# Create heatmap using matplotlib
fig, ax = plt.subplots(figsize=(12, 10))

# Create the heatmap with custom colormap and normalization
im = ax.imshow(tm_align_scores, cmap=custom_cmap, norm=divnorm, aspect='auto')
score_CIRPIN_filt = score_cirpin[0::2, 1::2]

#im = ax.imshow(score_CIRPIN_filt, cmap='RdBu_r', norm=divnorm, aspect='auto')

# Set ticks and labels
ax.set_xticks(np.arange(len(x_axis_ids)))
ax.set_yticks(np.arange(len(y_axis_ids)))
ax.set_xticklabels(x_axis_ids, fontsize=20)
ax.set_yticklabels(y_axis_ids, fontsize=20)

# Rotate the x-axis labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
plt.setp(ax.get_yticklabels(), rotation=45, ha="right")

# Add colorbar
cbar = plt.colorbar(im, ax=ax, aspect=15)
cbar.ax.tick_params(labelsize=20)
#cbar.set_label('Min TM-score', rotation=270, labelpad=20, fontsize=18)

# Set labels and title
ax.set_title('TM-align -cp scores', fontsize=36, pad=20)

plt.tight_layout()

# Save the figure
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
print("\nHeatmap saved")

# Save scores to numpy file
#np.save('tm_scores.npy', scores)
#print("Scores saved as 'tm_scores.npy'")

# Print summary statistics
# print("\nSummary Statistics:")
# print(f"Mean TM-score: {np.nanmean(scores):.3f}")
# print(f"Median TM-score: {np.nanmedian(scores):.3f}")
# print(f"Min TM-score: {np.nanmin(scores):.3f}")
# print(f"Max TM-score: {np.nanmax(scores):.3f}")

plt.show()
```
